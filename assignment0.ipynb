```python
print("performing a linear regression of the form Zx = y:")
Z = np.array([[1, 2],
              [3, 4],
              [5, 6]])
y = np.array([7, 8, 9])
x, residuals, rank, s = linalg.lstsq(Z, y)
print("Solution vector x:\n", x)
print("\nResiduals:\n", residuals)
print("\nRank of matrix Z:\n", rank)
print("\nSingular values of Z:\n", s)
```

```python
x = np.linspace(0, 10 * np.pi, 100)  
y = np.sin(x) 
q = 2

y_resampled = signal.resample(y, int(np.ceil(len(y) / q)))
print("downsample with low-pass filtering:")
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Original Signal', marker='o')
plt.plot(np.linspace(0, 10 * np.pi, len(y_resampled)), y_resampled, label='Resampled Signal', marker='x')
plt.title('Original vs Resampled Signal')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.random.rand(50)
y = 2.5 * x + np.random.randn(50) * 0.2  # Linear relationship with noise

# Perform linear regression
slope, intercept, r_value, p_value, std_err = linregress(x, y)
line = slope * x + intercept

# Create the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', label='Data Points')

# Plot the regression line
plt.plot(x, line, color='red', linewidth=2, label=f'Fit Line: y={slope:.2f}x + {intercept:.2f}')

# Add titles and labels
plt.title('Scatter Plot with Regression Line')
plt.xlabel('X Values')
plt.ylabel('Y Values')

# Add legend and grid
plt.legend()
plt.grid(True)

# Display the plot
plt.show()
print("\n")
print("END OF TASK 4!\n")
print("--------------------\n")
```
